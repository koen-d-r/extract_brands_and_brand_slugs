```python
# import libraries
import re
import requests
from bs4 import BeautifulSoup
```

### Extract available brands from clients' page

In this notebook I'm extracting all available brands and categories from a website by scraping with Beautiful Soup.

```python
# get URL and put in soup
url = 'https://www.schuurman-schoenen.nl/merken'
response = requests.get(url)
soup = BeautifulSoup(response.content)
```

```python
# extract brands and brand slugs from HTML
brand_slug_dict = {}

for link in soup.findAll('a', {'class': 'brand'}):
    match = re.search('href="(.*)">(.*)</a>', str(link))
    if match:
        brand_slug_dict[match.group(1)] = match.group(2)
```

```python
brand_slug_dict
```

### Extract categories from homepage menu

```python
# grab homepage for extracting categories
hp = requests.get('https://www.schuurman-schoenen.nl/')

# throw in soup
hp_soup = BeautifulSoup(hp.content)
```

```python
# create category dictionary with brand slug and category
cat_dict = {}

for link in hp_soup.findAll('a', {'class': 'nav-level-2-link nav-link js-nav-link nav-link--no-children d-lg-inline-block py-lg-1'}):
    match = re.search('href=\"(.*?)\".+>(.*)<\/a>', str(link))
    if match:
        cat_dict[match.group(1)] = match.group(2)
```
